```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
dataset1 = pd.read_csv("spy-put-call-fear-greed-vix.csv")
dataset1
```

```python
dataset2 = pd.read_csv("spy_vix_fear_greed_2011_2023.csv")
dataset2
```

```python
#Open (first dataset) and Open_SPY (second dataset) both refer to the opening price of stocks.
#High (first dataset) and High_SPY (second dataset) both refer to the highest price of stocks.
#Low (first dataset) and Low_SPY (second dataset) both refer to the lowest price of stocks.
#Close (first dataset) and Close_SPY (second dataset) both refer to the closing price of stocks.
#Volume (first dataset) and Volume_SPY (second dataset) both refer to the trading volume of stocks.

#Therefore we can drop repeated columns

dataset1.drop(columns=['Open','High', 'Low', 'Close', 'Volume'], inplace=True)
#And also FearGreed index is repeated too! so we drop it as well
dataset1.drop(columns= ['Fear Greed'], inplace=True)
dataset1
```

```python
df = dataset1.merge(dataset2, right_on = 'Date', left_on = 'Date')
df
```

Here's what each term stands for in the dataset:

Open_VIX: The opening value of the VIX (Volatility Index) for the trading day.

High_VIX: The highest value the VIX reached during the trading day.

Low_VIX: The lowest value the VIX reached during the trading day.

Close_VIX: The closing value of the VIX for the trading day.

Adj Close_VIX: The adjusted closing value of the VIX, accounting for any corporate actions like dividends or stock splits.

Open_SPY: The opening value of the SPDR S&P 500 ETF (SPY) for the trading day.

High_SPY: The highest value the SPY reached during the trading day.

Low_SPY: The lowest value the SPY reached during the trading day.

Close_SPY: The closing value of the SPY for the trading day.

Adj Close_SPY: The adjusted closing value of the SPY, accounting for any corporate actions like dividends or stock splits.

Volume_SPY: The trading volume of the SPY for the trading day.
